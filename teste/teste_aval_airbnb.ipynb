```python

# Carregue os dados do arquivo listings.csv.gz
file_path = "listings1.csv"
df = pd.read_csv(file_path)

# Visualize as primeiras linhas do dataframe
print(df.head())

# Obtenha estatísticas descritivas para variáveis numéricas
numeric_stats = df.describe()

# Explore a distribuição de variáveis categóricas
categorical_counts = df['property_type'].value_counts()

# Verifique a quantidade de dados ausentes em cada coluna
missing_data = df.isnull().sum()

# Imprima os resultados
print("\nEstatísticas Descritivas:")
print(numeric_stats)

print("\nDistribuição de Tipos de Propriedade:")
print(categorical_counts)

print("\nDados Ausentes:")
print(missing_data)
```
