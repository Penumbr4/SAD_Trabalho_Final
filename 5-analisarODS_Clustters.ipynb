Obter um dataframe só com os indicadores (indexados aos países)

```python
# Ficar só com as colunas dos indicadores e nomear as linhas pelo país
indicadores2022 = ODS2022.iloc[:, 4:].set_index(ODS2022['Country'])
indicadores2022.head()
```

ANÁLISE HIERÁRQUICA DE CLUSTERS

Normalizar os dados (substrair pela média e dividir pelo desvio padrão)

```python
import sklearn
```

```python
from sklearn.preprocessing import StandardScaler

# Normalização dos dados usando StandardScaler
scaler = StandardScaler()
indicadores_scaled = scaler.fit_transform(indicadores2022)
print(indicadores_scaled)
```

```python
indicadores_scaled.shape
```

Obter a matriz de dissemelhanças e o dendograma

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# Obter a matriz de dissemelhanças
linkage_matrix = linkage(indicadores_scaled, method='ward',metric='euclidean')

# Criar o dendograma
plt.figure(figsize=(12, 8))
dendrogram(linkage_matrix, orientation='top', distance_sort='descending',labels=indicadores2022.index,show_leaf_counts=True)
plt.title('Dendograma de Análise Hierárquica de Clusters')
plt.xlabel('Distância Euclideana')
plt.show()
```

Obter a solução para um determinado número de clusters

```python
from scipy.cluster.hierarchy import cut_tree

# Usar cut_tree para cortar o dendograma e obter os rótulos de cluster. Colocar em n_clusters o nºclusters pretendido.
cluster_labels4 = cut_tree(linkage_matrix, n_clusters=4)
```

Obter o número de observações por cluster

```python
import numpy as np

cluster_labels4 = cluster_labels4.flatten() #converte a matriz de rótulos numa lista
count= np.bincount(cluster_labels4) #conta o número de observações em cada cluster
print(count)
```

Acrescentar uma coluna com o cluster no DataFrame

```python
# Adicionar a coluna 'ClusterH4' com os rótulos dos clusters ao dataframe ODS2022
ODS2022.loc[:, 'ClusterH4'] = cluster_labels4
ODS2022.head()
```

```python
ODS2022['ClusterH4'].value_counts()
```

Obter a lista de países em cada cluster

```python
clustersH4 = ODS2022.groupby('ClusterH4')['Country'].apply(list) 
for cluster, countries in clustersH4.items():
    print(f'Cluster {cluster}: {", ".join(countries)}')
```

Obter as médias dos indicadores nos clusters (centróides)

```python
indicadores = ['SDG1 ', 'SDG2 ', 'SDG3 ', 'SDG4 ', 'SDG5 ', 'SDG6 ', 'SDG7 ', 'SDG8 ', 'SDG9 ', 'SDG10 ',
               'SDG11 ', 'SDG12 ', 'SDG13 ', 'SDG14 ', 'SDG15 ', 'SDG16 ', 'SDG17 ']
médias_cluster = ODS2022.groupby('ClusterH4')[indicadores].mean()
print(médias_cluster)
```

ANÁLISE NÃO HIERÁRQUICA DE CLUSTERS

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4,random_state=42,n_init=10)

#n_clusters=4: Número de clusters

#random_state=42: Semente (seed) para gerar números aleatórios. 
#n_init=10: O k-means é inicializado 10 vezes e é escolhida a melhor solução
#a melhor é a solução com menor inércia (soma dos quadrados das distâncias das observações aos seus centróides) 

kmeans.fit(indicadores_scaled)

# Obter os rótulos dos clusters para cada observação
# cluster_labels_km = kmeans.labels_    #_indica que a variável é um atributo calculado por um modelo de machine learning
# ou
cluster_labels_km4 = kmeans.fit_predict(indicadores_scaled)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4,random_state=42,n_init=10)
kmeans.fit(indicadores_scaled)
cluster_labels_km4 = kmeans.fit_predict(indicadores_scaled)
```

```python
ODS2022.loc[:, 'ClusterKM4'] = cluster_labels_km4
ODS2022.head()
```

```python
ODS2022['ClusterKM4'].value_counts().sort_index()
```

```python
cluster_counts = ODS2022['ClusterKM4'].value_counts().sort_index()
# Crie um gráfico de barras para mostrar a distribuição de observações por cluster
cluster_counts.plot(kind='bar')
plt.xlabel('Cluster')
plt.ylabel('Número de Países')
plt.title('Distribuição de Países por Cluster')
plt.show()
```

Obter a lista de países em cada cluster

```python
clusters = ODS2022.groupby('ClusterKM4')['Country'].apply(list) 
for cluster, countries in clusters.items():
    print(f'Cluster {cluster}: {", ".join(countries)}')
```

COMO ESCOLHER O NÚMERO MAIS ADEQUADO DE CLUSTERS?

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

k_values = range(2, 11)
silhouette_scores = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(indicadores_scaled)
    silhouette_avg = silhouette_score(indicadores_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

best_k = k_values[silhouette_scores.index(max(silhouette_scores))] #encontra o k que maximiza o índice de silhueta

plt.plot(k_values, silhouette_scores, marker='o')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Média da Silhueta')
plt.title('Método da Silhueta para Escolha do Número de Clusters')
plt.show()
print(f'O número recomendado de clusters (k) é: {best_k}')
```

Gráfico de Elbow

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia_values = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(indicadores_scaled)
    inertia_values.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia_values, marker='o')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inércia')
plt.title('Gráfico de Elbow')
plt.show()
```

Solução com 2 clusters obtidos pelo método das k médias

```python
kmeans = KMeans(n_clusters=2,random_state=42,n_init=10)
kmeans.fit(indicadores_scaled)
cluster_labels_km2 = kmeans.fit_predict(indicadores_scaled)

ODS2022.loc[:, 'ClusterKM2'] = cluster_labels_km2
contagens = ODS2022['ClusterKM2'].value_counts().sort_index()
print(contagens)

clusters = ODS2022.groupby('ClusterKM2')['Country'].apply(list) 
for cluster, countries in clusters.items():
    print(f'Cluster {cluster}: {", ".join(countries)}')
médias_clusterKM2 = ODS2022.groupby('ClusterKM2')[indicadores].mean()
print(médias_clusterKM2)
```

Representar graficamente os clusters

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Opcional, mas pode melhorar a estética dos gráficos
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Aplicar Análise de Componentes Principais para reduzir para duas dimensões
pca = PCA(n_components=2)
indicadores_pca = pca.fit_transform(indicadores_scaled)

# Representar graficamente, com cores diferentes por cluster
plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(x=indicadores_pca[:, 0], y=indicadores_pca[:, 1],
                           hue=ODS2022['ClusterKM2'], palette='viridis', legend='full')
# Adicionar os nomes dos países como rótulos dos pontos
for i, país in enumerate(ODS2022['Country']):
    scatter.text(indicadores_pca[i, 0], indicadores_pca[i, 1], país, fontsize=8, ha='left', va='bottom')
plt.title('Diagrama de dispersão das Componentes Principais dos indicadores')
plt.show()
```

Tentativa de representar no mapa da Europa... (a melhorar)

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
#pode-se obter em https://www.naturalearthdata.com/downloads/110m-cultural-vectors/

europe = world[world['continent'] == 'Europe']

europe_clusters = europe.merge(ODS2022[['Country Code ISO3', 'ClusterKM2']], left_on='iso_a3', right_on='Country Code ISO3')

fig, ax = plt.subplots(1, 1, figsize=(10, 10))

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)

europe_clusters.plot(column='ClusterKM2', cmap='viridis', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True, cax=cax)
plt.title('Mapa da Europa com Coloração por Cluster')
plt.show()
```
